```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

```python
df = spark.read.json("transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

```python
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

```python
from pyspark.sql.functions import min as _min, max as _max, sum as _sum, round as _round

category_summary = (
    df.groupBy("category")
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(_min("amount"), 2).alias("min_PLN"),
        _round(_max("amount"), 2).alias("max_PLN")
    )
    .orderBy("category")
)


category_summary.show()
```

```python
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

```python
from pyspark.sql.functions import window, count, sum as _sum, round as _round

# Grupujemy po 30-minutowym oknie oraz po sklepie
windowed_store_summary = (
    df.groupBy(window("timestamp", "30 minutes"), "store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy("window", "store")
)

windowed_store_summary.show(truncate=False)
```

```python
from pyspark.sql.functions import desc, window, sum as _sum, round as _round

krakow_top_hour = (
    df.filter(df.store == "Kraków") 
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy(desc("suma_PLN"))
)

krakow_top_hour.show(truncate=False)
```

```python
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)
sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)
print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")

# Odpowiedz w komentarzu: dlaczego sliding ma więcej wierszy?
# TWOJA ODPOWIEDŹ: W oknie tumbling każda transakcja jest przypisywana do dokładnie jednego, unikalnego przedziału czasu. 
# W oknie sliding (o szerokości 1h i kroku 30 minut) przedziały czasowe nakładają się na siebie.
```

```python
# Odpowiedz na pytania w komentarzach:

# 1. Ile transakcji jest w oknie 09:00–10:00?
#    Sprawdź w wyniku zadania 3.1.
#    ODPOWIEDŹ: {2026-04-12 09:00:00, 2026-04-12 10:00:00}|4661

# 2. Jaka jest różnica między groupBy("store") a groupBy(window(...), "store")?
#    ODPOWIEDŹ: `groupBy("store")` bierze pod uwagę wszystkie transakcje danego sklepu. 
#    `groupBy(window(...), "store")` najpierw szatkuje oś czasu na kawałki (okna), 
#    a następnie liczy statystyki dla każdego sklepu niezależnie w każdym z tych kawałków.

# 3. W oknie sliding 1h/30min — ile okien zawiera transakcje z godziny 09:30?
#    Wskazówka: narysuj oś czasu.
#    ODPOWIEDŹ: Dokładnie 2 okna. W Sparku okna są domyślnie lewostronnie domknięte 
#    i prawostronnie otwarte [start, koniec). Transakcja ze ścisłej godziny 09:30:00 
#    wpadnie do okna [09:00–10:00) oraz do okna [09:30–10:30). Nie wpadnie natomiast 
#    do okna [08:30–09:30), ponieważ kończy się ono ułamek sekundy przed 09:30:00.
```

```python
from pyspark.sql.functions import avg, desc, asc
gdansk_lowest_avg = (
    df.filter(df.store == "Gdańsk") 
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(avg("amount"), 2).alias("srednia_PLN")
    )
    .orderBy("srednia_PLN")
)
gdansk_lowest_avg.show(1, truncate=False)
```

```python
liczba_transakcji = (
    df.groupBy(window("timestamp", "30 minutes"), "category")
    .agg(count("tx_id").alias("liczba_tx"))
    .filter(col("window.start").cast("string").like("%09:00:00%"))
    .orderBy("category")
)
liczba_transakcji.show(truncate=False)
```

```python
okno_15 = (
    df.groupBy(window("timestamp", "15 minutes"))
    .agg(count("tx_id").alias("liczba_tx"))
    .orderBy(desc("liczba_tx"))
)

okno_15.show(1, truncate=False)
```
